```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Data/total_percent_df.csv')
```

**Question**: Top 10 địa điểm có sự thay đổi hàng đầu từ giai đoạn 2000-2010 sang giai đoạn 2010-2019

```python
# Tính top 10 
avg_data_value = data.groupby(['locationdesc', 'year'])['data_value'].mean().reset_index()
top_10_locations = avg_data_value.groupby('locationdesc')['data_value'].mean().nlargest(10).index
filtered_data = data[data['locationdesc'].isin(top_10_locations)]
```

```python
# Visualization 
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='data_value', hue='locationdesc', data=filtered_data)
plt.title('Top 10 địa điểm thay đổi data_value theo năm')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```

**Nhận xét:**
- Các khu vực top đầu ở giai đoạn 2000-2010 có sự giảm đáng kể vào giai đoạn 2010-2019, ví dụ như: Glascock, Poinsett,..
- Các khu vực top cuối ở giai đoạn 2000-2010 lại có sự gia tăng mạnh vào giai đoạn 2010-2019, ví dụ như: Avoyelles, Major,...  

**Lợi ích:**
- Nguyên cứu nguyên nhân, chính sách hỗ trợ tại các khu vực có sự giảm để áp dụng.
- Tìm hiểu lý do các khu vực gia tăng mạnh để có các biện pháp, chính sách hỗ trợ thích hợp.

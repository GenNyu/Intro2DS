```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/smoothed_rate_df.csv')
```

**Question:** Sự thay đổi về số ca theo năm đối với nhóm người trên 65 tuổi

```python
df_q1 = df[(df['sex'] == 'Overall') & (df['race/ethnicityage_group'] == 'Overall') & (df['age_group'] == 'Ages 65+ years')]
df_q1_gr = df_q1.groupby('year').agg({'data_value': 'mean'}).reset_index()
df_q1_gr
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='data_value', data=df_q1, marker='o', ci=None)
plt.xticks(df_q1['year'].unique(), rotation=90)
plt.xlabel('')
plt.ylabel('')
plt.show()
```

**Nhận xét:**
- Giá trị của data_value biến động qua các năm và có xu hướng tăng dần vào những năm gần đây.
- Giá trị của data_value giảm mạnh vào năm 2009 và hiện đang đặt đỉnh trong những năm gần đây.  

**Lợi ích:**
- Phân tích tình hình xu hướng những năm gần đây có những ảnh hưởng gì dẫn tới sự tăng mạnh của data_value.
- Tìm hiểu những lý do giúp các năm 2009, 2010 có data_value đạt đáy để áp dụng.
